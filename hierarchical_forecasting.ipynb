```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pyplot
from google.colab import drive
file_path=('/content/drive/MyDrive/Dmart project/dmart_unprocessed_veda.csv')
data=pd.read_csv(file_path,parse_dates=True).loc[:,['Date','Region','City','Counter']]
data['Country']='India'
selected =['Pune', 'Mumbai', 'Nashik', 'Nagpur', 'Ahmedabad', 'Surat', 'Gandhinagar', 'Vadodara', 'Bangalore', 'Mysore', 'Chennai', 'Madurai', 'Hyderabad', 'Coimbatore', 'Kochi', 'New Delhi', 'Shimla', 'Chandigarh', 'Dehradun', 'Jammu', 'Agra', 'Kanpur', 'Jaipur', 'Manali', 'Kolkata', 'Patna', 'Darjeeling', 'Ranchi']
data = data.loc[data['City'].isin(selected), :]
data = data.rename(columns={'Date': 'ds', 'Counter': 'y'})
data.head()
```

```python
import pandas as pd
import datetime

# Assuming your DataFrame is named 'data' and the date column is 'ds'
data['ds'] = pd.to_datetime(data['ds'], format='%m/%d/%Y')

start_date = pd.to_datetime('12/31/2022', format='%m/%d/%Y')
end_date = pd.to_datetime('12/31/2023', format='%m/%d/%Y')

train = data[(data['ds'] < start_date)]
valid = data[(data['ds'] >= start_date) & (data['ds'] <= end_date)]
```

```python
from datetime import datetime, timedelta
h = 13
# Generate future dates for the forecast horizon
forecast_dates = [end_date + timedelta(days=30.44 * i) for i in range(1, h+1)]
# Print the forecast dates
for date in forecast_dates:
    print(date.strftime('%m-%d-%y'))
```

```python
from statsforecast import StatsForecast
from statsforecast.models import HoltWinters
model = StatsForecast(models=[HoltWinters(season_length=13, error_type='A')],
                              freq='M', n_jobs=-1)
model.fit(train_agg)

```

```python
p = model.forecast(h=h, fitted=True)
p_fitted = model.forecast_fitted_values()
```

```python
from hierarchicalforecast.methods import BottomUp, TopDown, MinTrace, ERM, OptimalCombination
from hierarchicalforecast.core import HierarchicalReconciliation

reconcilers = [BottomUp(),
               TopDown(method='forecast_proportions'),
               TopDown(method='average_proportions'),
               TopDown(method='proportion_averages'),
               MinTrace(method='ols', nonnegative=True),
               MinTrace(method='wls_struct', nonnegative=True),
               MinTrace(method='wls_var', nonnegative=True),
               MinTrace(method='mint_shrink', nonnegative=True),
               OptimalCombination(method='ols', nonnegative=True),
               OptimalCombination(method='wls_struct', nonnegative=True),
               ERM(method='closed'),
               ERM(method='reg'),
               ERM(method='reg_bu'),
              ]

rec_model = HierarchicalReconciliation(reconcilers=reconcilers)

p_rec = rec_model.reconcile(Y_hat_df=p, Y_df=p_fitted, S=S_train, tags=tags)
```

```python
p_rec
```

```python
p_rec_ = p_rec.merge(valid_agg, on=['ds', 'unique_id'], how='left')
p_rec_['y'] = p_rec_['y'].fillna(0)
p_rec_ = p_rec_.reset_index(drop=False)

from sklearn.metrics import mean_squared_error

rmse = dict()
for model_ in p_rec.columns[1:]:
    rmse_ = mean_squared_error(p_rec_['y'].values, p_rec_[model_].values, squared=False)/1e3
    # get only the model name
    model__ = model_.split('/')[-1]
    rmse[model__] = rmse_
pd.DataFrame(rmse, index=['RMSE']).T.sort_values('RMSE')
```
